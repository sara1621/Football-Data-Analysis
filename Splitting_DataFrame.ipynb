# Python Code: DataFrame Splitting into Mini-DataFrames by Class
 I will be using data from StatsBomb, specifically selecting the Argentina versus France match in the 2022 World Cup.
 


```python
import pandas as pd
from mplsoccer import Sbopen
parser = Sbopen()
```

```python
match_id=3869685
#Adding on the last tactics id and formation for the team for each event
events=parser.event(match_id)[0]
events.loc[events.tactics_formation.notnull(), 'tactics_id']= events.loc[
   events.tactics_formation.notnull(), 'id']
events[['tactics_id', 'tactics_formation']] = events.groupby('team_name')[[
   'tactics_id', 'tactics_formation']].ffill()  

# Selecting only the columns required for my analysis.
events=events[['id','team_name','x',"y",'player_name','type_name', 'tactics_formation','under_pressure'
              ]]
```

# 
I will be subdividing my DataFrame based on the formations adopted during the match.

```python
# Grouping the data based on teams and formations during the match
groupes = events.groupby(['team_name','tactics_formation'])
#Create mini-DataFrames 
mini_dataframes = [groupe.reset_index(drop=True) for _, groupe in groupes]

```

```python
argentina_352=mini_dataframes[0]
argentina_433=mini_dataframes[1]
france_442=mini_dataframes[2]
france_4231=mini_dataframes[3]
```
