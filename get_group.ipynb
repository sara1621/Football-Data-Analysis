# Selecting a specific group from a dataframe
To select a specific group, we will be using the get_group method after applying groupby on a statsbomb DataFrame.

```python
# Load the competitions 

df_competition= parser.competition()
df_competition
```

```python
# Bundesliga 2015/2016 has competition id 9, and season id 27
#Load the list of matches for this competition

matches=parser.match(competition_id=9,season_id=27)
matches
```

```python
#Look inside matches, I'll take only the needed columns 

matches=matches[['home_team_name','away_team_name','home_score','away_score','match_id','match_date']]
matches
```

```python
# I will choose the match  Borussia Dortmund VS Borussia Mönchengladbach 
# The ID of this match is 3890261.

# we open the data event [0]
data=parser.event(3890261)[0]
data
```

```python
# Group the DataFrame based on the 'type_name' column.

grouped_data=data.groupby("type_name")

#Determining the count of groups.

grouped_data.ngroups
```

```python
# We have 27 groups (27 Types)
# I will select the group of shots
# type_name== Shot

selected_group=grouped_data.get_group('Shot')

# Selecting only the important columns on the selected group

selected_group[["x",'y','end_x','end_y','player_name','match_id',"id","type_name","outcome_name","minute",'second',"under_pressure",'shot_statsbomb_xg']]
```
